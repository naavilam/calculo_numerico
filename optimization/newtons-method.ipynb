```python
def newton_method(f, f_prime, x0, tol=1e-6, max_iter=100):
    """
    Encontra a raiz de uma função usando o Método de Newton.

    Parâmetros:
        f (function): A função cujo zero queremos encontrar.
        f_prime (function): A derivada de f.
        x0 (float): Estimativa inicial.
        tol (float): Tolerância para o erro (critério de parada).
        max_iter (int): Número máximo de iterações.

    Retorna:
        float: Aproximação da raiz.
        int: Número de iterações realizadas.
    """
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        
        if abs(fpx) < 1e-12:  # Evita divisão por zero
            raise ValueError("Derivada próxima de zero. O método pode não convergir.")
        
        x_new = x - fx / fpx
        
        if abs(x_new - x) < tol:  # Critério de convergência
            return x_new, i + 1
        
        x = x_new

    raise ValueError("Número máximo de iterações atingido. O método não convergiu.")

```

```python
# Definir a função e sua derivada
def f(x):
    return x**2 - 2

def f_prime(x):
    return 2 * x

# Parâmetros
x0 = 1.0  # Estimativa inicial
tol = 1e-6  # Tolerância
max_iter = 100  # Número máximo de iterações

# Encontrar a raiz
raiz, iteracoes = newton_method(f, f_prime, x0, tol, max_iter)

print(f"Raiz encontrada: {raiz:.6f} em {iteracoes} iterações")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Geração de valores para f(x)
x_vals = np.linspace(0, 2, 500)
y_vals = f(x_vals)

# Plot da função
plt.figure(figsize=(8, 6))
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.plot(x_vals, y_vals, label="f(x) = x^2 - 2", color="blue")
plt.scatter([raiz], [0], color="red", label=f"Raiz: x = {raiz:.6f}")
plt.title("Método de Newton")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
