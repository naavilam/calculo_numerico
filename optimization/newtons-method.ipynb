```python
def newton_method(f, f_prime, x0, tol=1e-6, max_iter=100):
    """
    Encontra a raiz de uma funo usando o Mtodo de Newton.

    Parmetros:
        f (function): A funo cujo zero queremos encontrar.
        f_prime (function): A derivada de f.
        x0 (float): Estimativa inicial.
        tol (float): Tolerncia para o erro (critrio de parada).
        max_iter (int): Nmero mximo de iteraes.

    Retorna:
        float: Aproximao da raiz.
        int: Nmero de iteraes realizadas.
    """
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        
        if abs(fpx) < 1e-12:  # Evita diviso por zero
            raise ValueError("Derivada prxima de zero. O mtodo pode no convergir.")
        
        x_new = x - fx / fpx
        
        if abs(x_new - x) < tol:  # Critrio de convergncia
            return x_new, i + 1
        
        x = x_new

    raise ValueError("Nmero mximo de iteraes atingido. O mtodo no convergiu.")

```

```python
# Definir a funo e sua derivada
def f(x):
    return x**2 - 2

def f_prime(x):
    return 2 * x

# Parmetros
x0 = 1.0  # Estimativa inicial
tol = 1e-6  # Tolerncia
max_iter = 100  # Nmero mximo de iteraes

# Encontrar a raiz
raiz, iteracoes = newton_method(f, f_prime, x0, tol, max_iter)

print(f"Raiz encontrada: {raiz:.6f} em {iteracoes} iteraes")

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerao de valores para f(x)
x_vals = np.linspace(0, 2, 500)
y_vals = f(x_vals)

# Plot da funo
plt.figure(figsize=(8, 6))
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.plot(x_vals, y_vals, label="f(x) = x^2 - 2", color="blue")
plt.scatter([raiz], [0], color="red", label=f"Raiz: x = {raiz:.6f}")
plt.title("Mtodo de Newton")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
