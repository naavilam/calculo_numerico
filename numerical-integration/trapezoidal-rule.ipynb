# Regra do Trapézio

A **Regra do Trapézio** é um método de integração numérica que aproxima a integral definida de uma função \( f(x) \) no intervalo \([a, b]\). Ela é baseada na ideia de dividir o intervalo em pequenos subintervalos e calcular a área de trapézios sob a curva.

## Fórmula




A aproximação é dada por:

\[
I \approx \frac{\Delta x}{2} \left[ f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n) \right],
\]

onde:
- \( \Delta x = \frac{b-a}{n} \) é o tamanho de cada subintervalo,
- \( n \) é o número de subintervalos,
- \( x_0, x_1, \dots, x_n \) são os pontos que dividem o intervalo.

## Passos
1. Divida o intervalo \([a, b]\) em \( n \) subintervalos de tamanho \( \Delta x \).
2. Calcule os valores da função \( f(x) \) nos pontos \( x_0, x_1, \dots, x_n \).
3. Substitua os valores na fórmula da Regra do Trapézio.

## Exemplo de Aplicação
Vamos calcular a integral:

\[
\int_0^\pi \sin(x) \, dx
\]

usando \( n = 6 \) subintervalos. A Regra do Trapézio divide o intervalo em pequenos trapézios, e a soma das áreas fornece uma aproximação da integral.

## Vantagens
- Simples de implementar.
- Melhor precisão do que métodos mais simples, como a Regra do Retângulo.

## Desvantagens
- Pode ser impreciso para funções com comportamento muito variável dentro de cada subintervalo.
- A precisão depende do número de subintervalos \( n \).

## Complexidade
O método tem complexidade \( O(n) \), onde \( n \) é o número de subintervalos.

## Visualização
A aproximação pode ser representada graficamente pela soma das áreas dos trapézios formados entre os pontos \( (x_i, f(x_i)) \) e \( (x_{i+1}, f(x_{i+1})) \).


```python
def trapezoidal_rule(f, a, b, n):
    """
    Aproxima a integral definida de f(x) no intervalo [a, b] usando a Regra do Trapézio.

    Parâmetros:
        f (function): Função a ser integrada.
        a (float): Limite inferior de integração.
        b (float): Limite superior de integração.
        n (int): Número de subintervalos.

    Retorna:
        float: Aproximação da integral.
    """
    x = [a + i * (b - a) / n for i in range(n + 1)]  # Pontos de partição
    y = [f(xi) for xi in x]  # Valores da função nos pontos
    h = (b - a) / n  # Tamanho do subintervalo
    integral = (h / 2) * (y[0] + 2 * sum(y[1:-1]) + y[-1])
    return integral
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_trapezoidal(f, a, b, n):
    """
    Visualiza a aproximação da integral usando a Regra do Trapézio.

    Parâmetros:
        f (function): Função a ser integrada.
        a (float): Limite inferior de integração.
        b (float): Limite superior de integração.
        n (int): Número de subintervalos.
    """
    x = np.linspace(a, b, n + 1)
    y = f(x)
    x_fine = np.linspace(a, b, 500)
    y_fine = f(x_fine)

    # Plot da função original
    plt.figure(figsize=(8, 6))
    plt.plot(x_fine, y_fine, label="f(x)", color="blue")
    
    # Plot dos trapézios
    for i in range(n):
        plt.fill_between([x[i], x[i+1]], [y[i], y[i+1]], color='skyblue', alpha=0.4)

    plt.title("Regra do Trapézio: Aproximação")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import numpy as np

# Definir a função sinc(x) como sin(x)/x
def sinc(x):
    return np.where(x == 0, 1, np.sin(x) / x)


# Integração usando a Regra do Trapézio
a = 0.1  # Limite inferior (evite x=0 para evitar divisão por zero)
b = 10   # Limite superior
n = 8  # Número de subintervalos

# Aproximar a integral
# Calcular a integral com trapezoidal_rule
resultado = trapezoidal_rule(sinc, a, b, n)

# Exibir o resultado no console
print(f"A integral aproximada de sinc(x) no intervalo [{a}, {b}] é: {resultado:.4f}")

# Visualizar o gráfico com visualize_trapezoidal
visualize_trapezoidal(sinc, a, b, n)

```
