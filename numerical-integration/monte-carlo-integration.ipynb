```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_integration(f, a, b, N):
    """
    Calcula a integral de f(x) no intervalo [a, b] usando o método de Monte Carlo.

    Parâmetros:
        f (function): Função a ser integrada.
        a (float): Limite inferior de integração.
        b (float): Limite superior de integração.
        N (int): Número de amostras.

    Retorna:
        float: Aproximação da integral.
    """
    # Gerar N amostras uniformes no intervalo [a, b]
    x_samples = np.random.uniform(a, b, N)
    
    # Avaliar a função em cada amostra
    f_values = f(x_samples)
    
    # Calcular o valor médio de f(x)
    f_mean = np.mean(f_values)
    
    # Aproximar a integral
    integral = (b - a) * f_mean
    return integral
```

```python

# Função a ser integrada
f = np.sin

# Limites de integração
a = 0
b = np.pi

# Número de amostras
N = 100

# Calcular a integral usando Monte Carlo
resultado = monte_carlo_integration(f, a, b, N)

# Valor exato da integral
valor_exato = 2  # Sabemos que ∫₀^π sin(x) dx = 2

# Calcular o erro absoluto e relativo
erro_absoluto = abs(resultado - valor_exato)
erro_relativo = erro_absoluto / abs(valor_exato)

# Exibir resultados
print(f"Integral aproximada (Monte Carlo): {resultado:.6f}")
print(f"Valor exato: {valor_exato}")
print(f"Erro absoluto: {erro_absoluto:.6f}")
print(f"Erro relativo: {erro_relativo:.6%}")

# Visualizar as amostras e a curva
x_samples = np.random.uniform(a, b, N)
f_samples = f(x_samples)

x_fine = np.linspace(a, b, 1000)
y_fine = f(x_fine)

plt.figure(figsize=(8, 6))
plt.plot(x_fine, y_fine, label="f(x) = sin(x)", color="blue")
plt.scatter(x_samples, f_samples, s=1, alpha=0.5, color="red", label="Amostras")
plt.title("Integração de Monte Carlo")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
